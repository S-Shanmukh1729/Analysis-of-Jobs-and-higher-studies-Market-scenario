```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('response.csv') ## responses from google form
df.head()
```

```python
choices={}
for x in df['Carrer Choice']:
    choices[x]=choices.get(x,0)+1
```

```python
plt.figure(figsize=(7,7))
choice=list(choices.keys())                ## available choices after 12 and after
values=list(choices.values())    
percent=[i*100/sum(values) for i in values]     ## percent of people 
color=['Red','Blue','Yellow']
explode=[0,0,0.2]


plt.pie(values,labels=choice,colors=color,explode=explode,shadow=True)
plt.title("Stastics of carrer choices")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(choice,percent,color=color)
plt.title("Percentage of students pursuing different careers")
plt.show()


for i in range(len(choice)):
    print(f"People chosing {choice[i]} as the after choice of degree are {round(percent[i],2)} %")
```

## Students prefering Jobs analysis

```python
job=df.loc[df['Carrer Choice']=="Employment"]
job.head()
```

```python
job_count={'Data Science and Machine Learning':0,"Business Development (Sales)":0,"Software Development":0,"Data Analyst":0,"Product Management":0,"UI UX Design":0,"Other":0}
for i in job['field']:
    if i in job_count.keys():
        job_count[i]=job_count.get(i,0)+1
    else:
        job_count['Other']+=1
        
print(job_count)
```

```python
jobs_choi=list(job_count.keys())
job_percent=[round(i*100/sum(job_count.values()),3) for i in job_count.values()]

plt.figure(figsize=(23,10))
plt.bar(jobs_choi,job_percent)
plt.title("Statistics of students choosing job as their career")
for index, value in enumerate(job_percent):
    plt.text( index, value,str(value))
```

```python
plt.figure(figsize=(10,7))

def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n".format(pct)

plt.pie(job_percent,labels=jobs_choi,autopct = lambda pct: func(pct, job_percent),shadow=True)
plt.show()
```

## For students intrested in Jobs

We have a total of 2.45 lakh students intrested in doing jobs
Lets find how many persons are intrested in doing jobs in specific fields

```python
job_num={}
for i in range(len(job_percent)):
    print(f"There are {job_percent[i]}% people are intrested in doing {jobs_choi[i]} jobs .So a total of {round(job_percent[i]*2.45/100,3)} Lakh students are intrested in doing {jobs_choi[i]} job\n")
    job_num[jobs_choi[i]]=int(round(job_percent[i]*2.45/100,3)*100000)
```

```python
print(job_num)
```

```python
names=list(jobs['Description'])
ava=list(jobs['Full Time'])
```

```python
ava_jobs={}
for i in range(len(names)):
    ava_jobs[names[i]]=ava[i]
    
print(ava_jobs)
```

```python
plt.figure(figsize=(25,10))

X_axis = np.arange(len(ava_jobs))
  
a=[ava_jobs[j] for j in sorted(ava_jobs)]
b=[job_num[j] for j in sorted(ava_jobs)]
    
plt.bar(X_axis - 0.2,a , 0.4, label = 'Available')
plt.bar(X_axis + 0.2,b, 0.4, label = 'Required')
  
plt.xticks(X_axis, sorted(ava_jobs))
plt.xlabel("Fields")
plt.ylabel("Number of jobs")
plt.title("Number of jobs required vs no of jobs available")
plt.legend()
plt.show()
```

### Report or Conclusion

As we can see jobs required and available are **almost same** in the field of **Business Development**.

Whereas Required jobs are  **less than available jobs** in **Product Managment**. So there are less students showing intrest towards Product Managment eventhough there are many jobs.

As for **all the other fields** available jobs are **much less than required** jobs.So there will be a heavy competetion for students to get that job.

We can see a **huge difference in the field Data Science and Machine Learning** . This is due to increase in recent popularity for data science.
